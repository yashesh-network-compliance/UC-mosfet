```python
charging = trinoFetch("""
SELECT
    deviceId,
    startLat,
    startLon,
    startTs,
    endTs
FROM iceberg.silver.charging_discharging_events_new
WHERE deviceId IS NOT NULL
  AND state = 'charging'
  AND startTs >= CURRENT_DATE - INTERVAL '3' DAY
  AND startTs <  CURRENT_DATE
  AND (endSoc - startSoc) > 10
  AND startLat <> 25
  AND startLon <> -71
""")

print(f"  Charging events fetched: {len(charging)}")
print(charging["startTs"].min())
print(charging["startTs"].max())
```

```python
charging_df=charging.copy()
```

```python

print("\nSTEP 2: Fetching reference locations...")

partners_df = prodFetch('''
SELECT id AS partnerId, zoneId, latitude AS p_lat, longitude AS p_long
FROM partners
WHERE deletedAt IS NULL AND status = 'active' AND deployedBatteries > 0;
''')

warehouses_df = prodFetch('''
SELECT id AS warehouseId,zoneId, latitude AS w_lat, longitude AS w_long
FROM warehouses
WHERE deletedAt IS NULL AND status = 'active';
''')

client_df = prodFetch('''
SELECT id AS clientId,zoneId, latitude AS c_lat, longitude AS c_long
FROM entities
WHERE id LIKE 'CH%' AND status = 'active' AND deletedAt IS NULL
''')


print(f"  Partners: {len(partners_df)} | Warehouses: {len(warehouses_df)} | "
      f"Clients: {len(client_df)}")
```

```python
print("\nSTEP 3: Building location BallTree index...")

partners_df["type"] = "partner"
partners_df = partners_df.rename(
    columns={
        "partnerId": "location_id",
        "p_lat": "lat",
        "p_long": "lon"
    }
)

warehouses_df["type"] = "warehouse"
warehouses_df = warehouses_df.rename(
    columns={
        "warehouseId": "location_id",
        "w_lat": "lat",
        "w_long": "lon"
    }
)

client_df["type"] = "client"
client_df = client_df.rename(
    columns={
        "clientId": "location_id",
        "c_lat": "lat",
        "c_long": "lon"
    }
)

# =========================================
# COMBINE ALL LOCATIONS
# =========================================

locations = pd.concat([
    partners_df[[
        "location_id",
        "lat",
        "lon",
        "type",
        "zoneId"
    ]],

    warehouses_df[[
        "location_id",
        "lat",
        "lon",
        "type",
        "zoneId"
    ]],

    client_df[[
        "location_id",
        "lat",
        "lon",
        "type",
        "zoneId"
    ]]

], ignore_index=True)

# =========================================
# RADIAN CONVERSION
# =========================================

locations["lat_rad"] = np.radians(locations["lat"])
locations["lon_rad"] = np.radians(locations["lon"])

charging_df["lat_rad"] = np.radians(charging_df["startLat"])
charging_df["lon_rad"] = np.radians(charging_df["startLon"])

# =========================================
# BALL TREE
# =========================================

tree = BallTree(
    locations[["lat_rad", "lon_rad"]].values,
    metric="haversine"
)

dist, ind = tree.query(
    charging_df[["lat_rad", "lon_rad"]].values,
    k=1
)

# =========================================
# TAGGING
# =========================================

charging_df["distance_m"] = (
    dist.flatten() * 6371000
)

charging_df["nearest_index"] = ind.flatten()

charging_df["location_type"] = charging_df[
    "nearest_index"
].map(
    locations["type"].to_dict()
)

charging_df["nearest_zone"] = charging_df[
    "nearest_index"
].map(
    locations["zoneId"].to_dict()
)

charging_df["charging_location_type"] = np.where(
    charging_df["distance_m"] <= 200,
    charging_df["location_type"],
    "outside"
)
```

```python
batteries = prodFetch(f""" select id as batteryId, serialNo, iotDeviceNo from batteries b where isBaas=0 """)


txnLogs = prodFetch(f"""
select t.partnerId,t.driverId, t.batteriesIssued, t.batteriesReceived,t.createdAt, t.updatedAt,
t.status,t.id as txnId from transactions t
where createdAt >= CURRENT_DATE() - interval 3 day
and t.deletedAt is  null """)

txnLogs['createdAt'] = txnLogs['createdAt']+pd.Timedelta(minutes=330)
txnLogs['updatedAt'] = txnLogs['updatedAt']+pd.Timedelta(minutes=330)


txnLogs['date'] = txnLogs['createdAt'].dt.date

import ast
txnLogs['batteriesReceived'] = txnLogs['batteriesReceived'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
txnLogs['batteriesIssued'] = txnLogs['batteriesIssued'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

txnLogs['batteryRc1'] = txnLogs['batteriesReceived'].apply(lambda x: x[0] if len(x) > 0 else None)
txnLogs['batteryRc2'] = txnLogs['batteriesReceived'].apply(lambda x: x[1] if len(x) > 1 else None)
txnLogs['batteryIs1'] = txnLogs['batteriesIssued'].apply(lambda x: x[0] if len(x) > 0 else None)
txnLogs['batteryIs2'] = txnLogs['batteriesIssued'].apply(lambda x: x[1] if len(x) > 1 else None)


txnLogs = txnLogs.melt(
    id_vars=['partnerId', 'driverId', 'createdAt','updatedAt','date'],
    value_vars=['batteryRc1', 'batteryRc2', 'batteryIs1', 'batteryIs2'],
    var_name='batteryType',
    value_name='batteryId'
)
```

```python
txnLogs2 = txnLogs.merge(batteries,on='batteryId',how='left')
txnLogs2.sort_values(by='createdAt',inplace=True)
txnLogs2=txnLogs2[txnLogs2['serialNo'].isna()==False]
txnLogs2
```
